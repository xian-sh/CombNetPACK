```python
import numpy as np

npz_path = "../data_demo/data.npz"

data = np.load(npz_path, allow_pickle=True)
print("Keys contained:", list(data.keys()))

for key in data.keys():
    arr = data[key]
    try:
        shape = arr.shape
    except Exception:
        shape = "unknown"
    
    print(f"\nKey: {key}")
    print(f"  Shape: {shape}")
    print(f"  Dtype: {getattr(arr, 'dtype', type(arr))}")
    
    # Print first three entries (if 1D or 2D, print by row; if object array, display with repr)
    try:
        sample = arr[:3]
    except Exception:
        # Some arrays might be scalar or not sliceable
        sample = arr
    
    print("  First 3 samples:")
    print(sample)
```
